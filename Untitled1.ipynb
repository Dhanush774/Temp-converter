## Data exploration

### Subtask:
Explore the dataset to understand its structure, identify missing values and potential outliers.


**Reasoning**:
Examine data types, summary statistics, missing values, unique values in categorical columns, and potential outliers in numerical columns.



```python
# Examine data types
print(df.dtypes)

# Generate descriptive statistics for numerical columns
print(df.describe())

# Identify missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("Missing Values:\n", missing_values)
print("\nMissing Value Percentage:\n", missing_percentage)

# Analyze unique values for categorical columns
for col in df.select_dtypes(include=['object']):
    print(f"\nUnique values in '{col}':\n{df[col].unique()}")
    print(f"\nValue counts in '{col}':\n{df[col].value_counts()}")

# Initial outlier detection (numerical columns)
# Focus on relevant numerical columns (e.g., Age, Income, IQ, etc.)
numerical_cols = ['Age', 'Annual Family Income ($)', 'IQ', 'Sleeping hours','Number of friends']
for col in numerical_cols:
    print(f"\nPotential outliers in '{col}':")
    print(df[col].describe())
    # Further analysis with box plots or histograms would be done in visualization steps
```

## Data cleaning

### Subtask:
Clean the data by handling missing values and outliers.


**Reasoning**:
Impute missing values for numerical and categorical columns, handle outliers in specified columns, and correct inconsistent values in 'Time spent playing outdoor sports' and 'Total Time spent working in front of screen'.



```python
# Impute missing values
for col in df.select_dtypes(include=['number']):
    df[col] = df[col].fillna(df[col].median())
for col in df.select_dtypes(include=['object']):
    df[col] = df[col].fillna(df[col].mode()[0])

# Outlier handling
cols_to_check = ['Age', 'Annual Family Income ($)', 'IQ', 'Sleeping hours', 'Number of friends']
for col in cols_to_check:
    # Winsorize outliers
    lower_limit = df[col].quantile(0.01)
    upper_limit = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

# Inconsistent value handling
df = df[df['Time spent playing outdoor sports'] >= 0]
df = df[df['Total Time spent working in front of screen'] >= 0]
```
